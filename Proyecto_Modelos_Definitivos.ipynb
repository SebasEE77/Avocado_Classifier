### Clasificación

#### Prueba usando EfficientNet-B3 para clasificación de Aguacates 4 Clases con Aumento de Datos

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
from time import time
import copy
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader, random_split, Subset
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from google.colab import drive
drive.mount('/content/drive')


BASE     = "/content/drive/MyDrive/Proyecto Aguacates"
DATA_DIR = f"{BASE}/Dataset CNN/Clasificación/4_Clases_Muestras"
CLASS_NAMES = ["Verde", "Pre-Maduro", "Maduro", "Podrido"]

NUM_CLASSES = 4
BATCH_SIZE  = 32

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Dispositivo de entrenamiento: {device}")

############################ DATA TRANSFORMER ############################
train_transform = transforms.Compose([

    transforms.Resize((320, 320)),
    transforms.CenterCrop((300, 300)),

    # Desplaza la imagen aleatoriamente en x hasta un 20%
    # Simula que el aguacate no está perfectamente centrado en la foto
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),

    transforms.RandomHorizontalFlip(p=0.3),

    transforms.RandomRotation(degrees=30),

    transforms.ColorJitter(
      brightness=0.2,
      contrast=0.2,
      saturation=0.2
    ),

    # Simula que la foto se tomó desde un ángulo ligeramente diferente al frontal
    transforms.RandomPerspective(distortion_scale=0.2, p=0.3),

    transforms.RandomApply([
        transforms.GaussianBlur(kernel_size=3)
    ], p=0.2),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),

])

# Transform para validación y test SIN aumento de datos
val_test_transform = transforms.Compose([
    transforms.Resize((320, 320)),
    transforms.CenterCrop((300, 300)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

############################ DIVISIÓN DE DATOS ############################
dataset = datasets.ImageFolder(root=DATA_DIR)

print(f"Total de imágenes cargadas : {len(dataset)}")
print(f"Clases detectadas          : {dataset.class_to_idx}")

# Calcular tamaños
n_total = len(dataset)
n_train = int(0.70 * n_total)
n_val   = int(0.15 * n_total)
n_test  = n_total - n_train - n_val

# Generar índices aleatorios reproducibles
indices       = torch.randperm(n_total, generator=torch.Generator().manual_seed(42)).tolist()
train_indices = indices[:n_train]
val_indices   = indices[n_train:n_train + n_val]
test_indices  = indices[n_train + n_val:]

# Crear DOS datasets separados con transforms diferentes
train_dataset    = datasets.ImageFolder(root=DATA_DIR, transform=train_transform)
val_test_dataset = datasets.ImageFolder(root=DATA_DIR, transform=val_test_transform)

# Aplicar índices a cada dataset
train_set = Subset(train_dataset, train_indices)
val_set   = Subset(val_test_dataset, val_indices)
test_set  = Subset(val_test_dataset, test_indices)

print(f"\nTrain : {len(train_set)} imágenes")
print(f"Val   : {len(val_set)}   imágenes")
print(f"Test  : {len(test_set)}  imágenes")

############################ DATA LOADER ############################
train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_set,   batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_set,  batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)

############################ DEFINICIÓN DEL MODELO ############################
# Cargar EfficientNet-B3 con pesos preentrenados en ImageNet
model = models.efficientnet_b3(weights=models.EfficientNet_B3_Weights.DEFAULT)

# En EfficientNet el clasificador se llama model.classifier (no model.fc)
# classifier[1] es la capa Linear final: entrada 1536 neuronas
in_features = model.classifier[1].in_features

# Reemplazar el clasificador: de 1000 clases (ImageNet) a 5 clases (madurez)
model.classifier = nn.Sequential(
    nn.Dropout(p=0.4),
    nn.Linear(in_features, 256),
    nn.ReLU(),
    nn.Dropout(p=0.2),
    nn.Linear(256, NUM_CLASSES)
)

model = model.to(device)

print(f"\nParámetros totales     : {sum(p.numel() for p in model.parameters()):,}")
print(f"Parámetros entrenables : {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

############################ FUNCIÓN DE PÉRDIDA ############################
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

############################ FUNCIÓN DE ENTRENAMIENTO ############################
def entrenar(model, criterion, optimizer, scheduler, train_loader, val_loader, num_epochs, patience=5):
    best_weights = copy.deepcopy(model.state_dict())
    best_val_acc = 0.0
    history = {"train_loss": [], "val_loss": [],
               "train_acc":  [], "val_acc":  []}

    no_improve = 0

    for epoch in range(num_epochs):
        start = time()

        ## FASE TRAIN
        model.train()
        train_loss, train_correct = 0.0, 0

        with tqdm(train_loader, desc=f"Época {epoch+1:02d}/{num_epochs} [Train]",
                  unit="batch", leave=False) as pbar:
            for images, labels in pbar:
                images, labels = images.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = model(images)
                loss    = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                _, preds       = torch.max(outputs, dim=1)
                train_loss    += loss.item() * images.size(0)
                train_correct += torch.sum(preds == labels).item()
                pbar.set_postfix(loss=f"{loss.item():.4f}")

        epoch_train_loss = train_loss    / len(train_set)
        epoch_train_acc  = train_correct / len(train_set)

        ## FASE VALIDACIÓN
        model.eval()
        val_loss, val_correct = 0.0, 0

        with torch.no_grad():
            with tqdm(val_loader, desc=f"Época {epoch+1:02d}/{num_epochs} [Val]  ",
                      unit="batch", leave=False) as pbar:
                for images, labels in pbar:
                    images, labels = images.to(device), labels.to(device)
                    outputs  = model(images)
                    loss     = criterion(outputs, labels)
                    _, preds = torch.max(outputs, dim=1)
                    val_loss    += loss.item() * images.size(0)
                    val_correct += torch.sum(preds == labels).item()

        epoch_val_loss = val_loss    / len(val_set)
        epoch_val_acc  = val_correct / len(val_set)

        if epoch_val_acc > best_val_acc:
            best_val_acc = epoch_val_acc
            best_weights = copy.deepcopy(model.state_dict())
            best_flag    = "  Mejor modelo"
            no_improve   = 0
        else:
            best_flag   = ""
            no_improve += 1
            if no_improve >= patience:
                print(f"\nEarly stopping activado en época {epoch+1}")
                break

        scheduler.step(epoch_val_acc)

        history["train_loss"].append(epoch_train_loss)
        history["val_loss"].append(epoch_val_loss)
        history["train_acc"].append(epoch_train_acc)
        history["val_acc"].append(epoch_val_acc)

        duration = (time() - start) / 60
        print(f"Época {epoch+1:02d}/{num_epochs} | "
              f"Tiempo: {duration:.1f} min | "
              f"Train Loss: {epoch_train_loss:.4f}  Acc: {epoch_train_acc:.4f} | "
              f"Val Loss: {epoch_val_loss:.4f}  Acc: {epoch_val_acc:.4f}"
              f"{best_flag}")

    print(f"\nEntrenamiento finalizado. Mejor val accuracy: {best_val_acc:.4f}")
    model.load_state_dict(best_weights)
    return model, history


############################ FINE-TUNING EN DOS FASES ############################

# FASE 1: Solo entrenar el clasificador
# Se congela todo el backbone (features). Solo model.classifier actualiza pesos.
for param in model.parameters():
    param.requires_grad = False
for param in model.classifier.parameters():
    param.requires_grad = True

print(f"\nFASE 1 — Parámetros entrenables: "
      f"{sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

optimizer_fase1 = optim.Adam(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=1e-3,
    weight_decay=1e-4
)

scheduler_fase1 = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer_fase1, mode='max', factor=0.5, patience=3
)

model, history_fase1 = entrenar(
    model, criterion, optimizer_fase1, scheduler_fase1,
    train_loader, val_loader,
    num_epochs=10, patience=4
)

# FASE 2: Descongelar todo y fine-tuning con LR muy bajo
# EfficientNet-B3 tiene más parámetros que ResNet-18, por eso se usa
# un LR aún más conservador (5e-5) para no destruir los pesos de ImageNet
for param in model.parameters():
    param.requires_grad = True

print(f"\nFASE 2 — Parámetros entrenables: "
      f"{sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

optimizer_fase2 = optim.Adam(
    model.parameters(),
    lr=1e-4,
    weight_decay=1e-4)

scheduler_fase2 = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer_fase2, mode='max', factor=0.5, patience=3
)

model, history_fase2 = entrenar(
    model, criterion, optimizer_fase2, scheduler_fase2,
    train_loader, val_loader,
    num_epochs=30, patience=8
)

torch.save(model.state_dict(), f"{BASE}/Dataset CNN/Clasificación/4_Clases/EfficientNet-B3/efficientnetb3_clasificacion.pth")
print(f"Modelo guardado en: {BASE}/Dataset CNN/Clasificación/4_Clases/EfficientNet-B3/efficientnetb3_clasificacion.pth")

############################ CURVAS DE ENTRENAMIENTO ############################
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_fase2["train_loss"], label="Train", color="#2196F3")
axes[0].plot(history_fase2["val_loss"],   label="Val",   color="#F44336")
axes[0].set_title("Función de Pérdida por Época")
axes[0].set_xlabel("Época")
axes[0].set_ylabel("Cross-Entropy Loss")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(history_fase2["train_acc"], label="Train", color="#2196F3")
axes[1].plot(history_fase2["val_acc"],   label="Val",   color="#F44336")
axes[1].set_title("Accuracy por Época")
axes[1].set_xlabel("Época")
axes[1].set_ylabel("Accuracy")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f"{BASE}/Dataset CNN/Clasificación/4_Clases/EfficientNet-B3/EfficientNetB3_curvas_entrenamiento_con_AD.png", dpi=150, bbox_inches="tight")
plt.show()

############################ EVALUACIÓN FINAL EN TEST SET ############################
model.eval()
all_preds  = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())

all_preds  = np.array(all_preds)
all_labels = np.array(all_labels)

accuracy  = accuracy_score(all_labels, all_preds)
precision = precision_score(all_labels, all_preds, average="weighted")
recall    = recall_score(all_labels, all_preds, average="weighted")
f1        = f1_score(all_labels, all_preds, average="weighted")

print("\n" + "="*55)
print(" MÉTRICAS DE DESEMPEÑO — TEST SET")
print("="*55)
print(f"  Accuracy  : {accuracy:.4f}  ({accuracy*100:.2f}%)")
print(f"  Precision : {precision:.4f}")
print(f"  Recall    : {recall:.4f}")
print(f"  F1-Score  : {f1:.4f}")
print("="*55)

print("\nReporte por clase:")
print(classification_report(all_labels, all_preds,
                             target_names=CLASS_NAMES, digits=4))

cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title("Matriz de Confusión — EfficientNet-B3 (Test Set)")
plt.ylabel("Etiqueta Real")
plt.xlabel("Predicción del Modelo")
plt.tight_layout()
plt.savefig(f"{BASE}/Dataset CNN/Clasificación/4_Clases/EfficientNet-B3/EfficientNetB3_matriz_confusion_con_AD.png", dpi=150, bbox_inches="tight")
plt.show()
print("Matriz de confusión guardada.")
```

### Regresión

#### Prueba usando EfficientNet-B3 para predecir la vida util de Aguacates con Aumento de Datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from time import time
import copy
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from PIL import Image

from google.colab import drive
drive.mount('/content/drive')

# Rutas del proyecto
BASE     = "/content/drive/MyDrive/Proyecto Aguacates"
CSV_PATH = f"{BASE}/Dataset CNN/Regresión/Dataset_Regresion_CNN.csv"
IMG_DIR  = f"{BASE}/Dataset CNN/Regresión/Imagenes"

BATCH_SIZE = 32

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Dispositivo de entrenamiento: {device}")

############################ CARGA DEL CSV ############################

df = pd.read_csv(CSV_PATH)

print(f"Total de imágenes : {len(df)}")
print(f"Días restantes — min: {df['dias_restantes'].min()}  "
      f"max: {df['dias_restantes'].max()}  "
      f"media: {df['dias_restantes'].mean():.2f}")

# Mezclar todas las filas aleatoriamente
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Calcular tamaños de cada subconjunto
n_total = len(df)
n_train = int(0.70 * n_total)  # 70% para entrenamiento
n_val   = int(0.15 * n_total)  # 15% para validación
n_test  = n_total - n_train - n_val  # 15% restante para test

# Dividir el dataframe en tres partes sin mezclarlas
df_train = df.iloc[:n_train].reset_index(drop=True)
df_val   = df.iloc[n_train:n_train + n_val].reset_index(drop=True)
df_test  = df.iloc[n_train + n_val:].reset_index(drop=True)

print(f"\nTrain : {len(df_train)} imágenes")
print(f"Val   : {len(df_val)}   imágenes")
print(f"Test  : {len(df_test)}  imágenes")

############################ DATA TRANSFORMER ############################
train_transform = transforms.Compose([

    transforms.Resize((320, 320)),
    transforms.CenterCrop((300, 300)),

    # Desplaza la imagen aleatoriamente en x hasta un 20%
    # Simula que el aguacate no está perfectamente centrado en la foto
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),

    transforms.RandomHorizontalFlip(p=0.3),

    transforms.RandomRotation(degrees=30),

    transforms.ColorJitter(
      brightness=0.2,
      contrast=0.2,
      saturation=0.2
    ),

    # Simula que la foto se tomó desde un ángulo ligeramente diferente al frontal
    transforms.RandomPerspective(distortion_scale=0.2, p=0.3),

    transforms.RandomApply([
        transforms.GaussianBlur(kernel_size=3)
    ], p=0.2),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),

])

# Transform para validación y test SIN aumento de datos
val_test_transform = transforms.Compose([
    transforms.Resize((320, 320)),
    transforms.CenterCrop((300, 300)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

############################ DATASET ############################
# PyTorch necesita una estructura que sepa cómo leer cada imagen del disco.
# Esta estructura (Dataset) hace tres cosas por cada imagen:
#   1. Leer el nombre del archivo desde el dataframe
#   2. Abrir la imagen del disco y aplicar el preprocesamiento
#   3. Retornar la imagen junto con su valor de días restantes
class DatasetRegresion(Dataset):

    # Recibe el dataframe, la carpeta de imágenes y el transform
    def __init__(self, df, img_dir, transform):
        self.df        = df         # dataframe
        self.img_dir   = img_dir    # ruta de la carpeta con las imágenes
        self.transform = transform  # preprocesamiento a aplicar

    # Retorna cuántas imágenes tiene este dataset
    def __len__(self):
        return len(self.df)

    # Se ejecuta cada vez que el DataLoader pide una imagen
    # idx es el índice de la imagen que se quiere leer
    def __getitem__(self, idx):
        nombre = self.df.loc[idx, "imagen"]
        dias   = self.df.loc[idx, "dias_restantes"]

        # Abrir la imagen del disco y convertirla a RGB
        img = Image.open(f"{self.img_dir}/{nombre}").convert("RGB")

        # Aplicar el preprocesamiento
        img = self.transform(img)

        # Retornar imagen procesada y días como tensor de tipo float
        return img, torch.tensor(dias, dtype=torch.float32)

train_set = DatasetRegresion(df_train, IMG_DIR, train_transform)
val_set   = DatasetRegresion(df_val,   IMG_DIR, val_test_transform)
test_set  = DatasetRegresion(df_test,  IMG_DIR, val_test_transform)

############################ DATA LOADER ############################
train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_set,   batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_set,  batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)

############################ DEFINICIÓN DEL MODELO ############################
# Cargar EfficientNet-B3 con pesos preentrenados en ImageNet
model = models.efficientnet_b3(weights=models.EfficientNet_B3_Weights.DEFAULT)

# En EfficientNet el clasificador se llama model.classifier (no model.fc)
# classifier[1] es la capa Linear final: entrada 1536 neuronas
in_features = model.classifier[1].in_features

model.classifier = nn.Sequential(
    nn.Dropout(p=0.4),           # desactiva 40% de neuronas para evitar sobreajuste
    nn.Linear(in_features, 256), # reduce de 512 a 256 neuronas
    nn.ReLU(),                   # función de activación no lineal
    nn.Dropout(p=0.2),           # segunda capa de regularización
    nn.Linear(256, 1)
)

model = model.to(device)

print(f"\nParámetros totales     : {sum(p.numel() for p in model.parameters()):,}")
print(f"Parámetros entrenables : {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

############################ FUNCIÓN DE PÉRDIDA ############################
# MSELoss: error cuadrático medio
# Mide cuánto difiere el número de días predicho del real
criterion = nn.MSELoss()

############################ MÉTRICAS DE REGRESIÓN ############################
def r2_adjusted(r2, n, v):
    return 1 - ((1 - r2) * (n - 1)) / (n - v - 1)

def smape(y_true, y_pred):
    return 100 * np.mean(
        2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred))
    )

def imprimir_metricas(y_true, y_pred, n_features):
    mae    = mean_absolute_error(y_true, y_pred)
    rmse   = np.sqrt(mean_squared_error(y_true, y_pred))
    smape_ = smape(y_true, y_pred)
    r2     = r2_score(y_true, y_pred)
    r2adj  = r2_adjusted(r2, len(y_true), n_features)

    print("\n" + "="*55)
    print("   MÉTRICAS DE DESEMPEÑO — TEST SET")
    print("="*55)
    print(f"  MAE         : {mae:.4f}  días")
    print(f"  SMAPE       : {smape_:.2f}%")
    print(f"  RMSE        : {rmse:.4f}  días")
    print(f"  R²          : {r2:.4f}")
    print(f"  R² ajustado : {r2adj:.4f}")
    print("="*55)

############################ FUNCIÓN DE ENTRENAMIENTO ############################

def entrenar(model, criterion, optimizer, scheduler, train_loader, val_loader,
             num_epochs, patience=5):

    # Guardar los mejores pesos encontrados durante el entrenamiento
    best_weights  = copy.deepcopy(model.state_dict())

    # En regresión se guarda el modelo con MENOR val loss
    best_val_loss = float("inf")

    history = {"train_loss": [], "val_loss": []}

    no_improve = 0

    for epoch in range(num_epochs):
        start = time()

        ## FASE TRAIN
        model.train()
        train_loss = 0.0

        with tqdm(train_loader, desc=f"Época {epoch+1:02d}/{num_epochs} [Train]",
                  unit="batch", leave=False) as pbar:
            for images, dias in pbar:
                images = images.to(device)
                dias   = dias.to(device).unsqueeze(1)       # (batch,) → (batch, 1) para MSELoss

                optimizer.zero_grad()                       # limpiar gradientes del paso anterior
                outputs = model(images)                     # predicción del modelo: (batch, 1)
                loss    = criterion(outputs, dias)          # calcular MSE entre predicción y real
                loss.backward()                             # calcular gradientes
                optimizer.step()                            # actualizar pesos

                train_loss += loss.item() * images.size(0) # acumular pérdida ponderada por batch
                pbar.set_postfix(loss=f"{loss.item():.4f}")

        # Promediar pérdida sobre todas las imágenes de train
        epoch_train_loss = train_loss / len(train_set)

        ## FASE VALIDACIÓN
        model.eval()
        val_loss = 0.0

        with torch.no_grad():   # no calcular gradientes en validación
            with tqdm(val_loader, desc=f"Época {epoch+1:02d}/{num_epochs} [Val]  ",
                      unit="batch", leave=False) as pbar:
                for images, dias in pbar:
                    images = images.to(device)
                    dias   = dias.to(device).unsqueeze(1)
                    loss   = criterion(model(images), dias)
                    val_loss += loss.item() * images.size(0)

        # Promediar pérdida sobre todas las imágenes de validación
        epoch_val_loss = val_loss / len(val_set)

        ## GUARDAR MEJOR MODELO
        # En regresión el mejor modelo es el que tiene MENOR val loss
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss          # actualizar mejor pérdida
            best_weights  = copy.deepcopy(model.state_dict())
            best_flag     = "  Mejor modelo"
            no_improve    = 0
        else:
            best_flag   = ""
            no_improve += 1
            if no_improve >= patience:
                print(f"\nEarly stopping activado en época {epoch+1}")
                break

        scheduler.step(epoch_val_loss)

        history["train_loss"].append(epoch_train_loss)
        history["val_loss"].append(epoch_val_loss)

        duration = (time() - start) / 60
        print(f"Época {epoch+1:02d}/{num_epochs} | "
              f"Tiempo: {duration:.1f} min | "
              f"Train Loss: {epoch_train_loss:.4f} | "
              f"Val Loss: {epoch_val_loss:.4f}"
              f"{best_flag}")

    print(f"\nEntrenamiento finalizado. Mejor val loss: {best_val_loss:.4f}")

    # Restaurar los mejores pesos encontrados durante el entrenamiento
    model.load_state_dict(best_weights)
    return model, history


############################ FINE-TUNING EN DOS FASES ############################

# FASE 1: Solo entrenar el regresor (fc)
# Congelar todos los parámetros del backbone — sus pesos no se actualizan
for param in model.parameters():
    param.requires_grad = False

# Descongelar SOLO el classifier (regresor)
for param in model.classifier.parameters():
    param.requires_grad = True

print(f"\nFASE 1 — Parámetros entrenables: "
      f"{sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

optimizer_fase1 = optim.Adam(
    model.classifier.parameters(),
    lr=1e-3,
    weight_decay=1e-4
)

scheduler_fase1 = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer_fase1, mode='min', factor=0.5, patience=3
)

model, history_fase1 = entrenar(
    model, criterion, optimizer_fase1, scheduler_fase1,
    train_loader, val_loader,
    num_epochs=10, patience=4
)

# FASE 2: Descongelar layer3, layer4 y fc
# Congelar todo de nuevo antes de descongelar selectivamente
for param in model.parameters():
    param.requires_grad = True

print(f"\nFASE 2 — Parámetros entrenables: "
      f"{sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

optimizer_fase2 = optim.Adam(
    model.parameters(),
    lr=1e-4,
    weight_decay=1e-4)

scheduler_fase2 = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer_fase2, mode='min', factor=0.5, patience=3
)

model, history_fase2 = entrenar(
    model, criterion, optimizer_fase2, scheduler_fase2,
    train_loader, val_loader,
    num_epochs=40, patience=8
)

torch.save(model.state_dict(), f"{BASE}/Dataset CNN/Regresión/EfficientNet-B3/efficientnetb3_regresion.pth")
print(f"Modelo guardado en: {BASE}/Dataset CNN/Regresión/EfficientNet-B3/efficientnetb3_regresion.pth")

############################ CURVAS DE ENTRENAMIENTO ############################

plt.figure(figsize=(8, 5))
plt.plot(history_fase2["train_loss"], label="Train", color="#2196F3")
plt.plot(history_fase2["val_loss"],   label="Val",   color="#F44336")
plt.title("Función de Pérdida por Época — EfficientNet-B3 Regresión")
plt.xlabel("Época")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(f"{BASE}/Dataset CNN/Regresión/EfficientNet-B3/Regresion_curvas_con_AD.png",
            dpi=150, bbox_inches="tight")
plt.show()

############################ EVALUACIÓN FINAL EN TEST SET ############################

model.eval()   # desactivar Dropout para evaluación final
all_preds  = []
all_labels = []

with torch.no_grad():
    for images, dias in test_loader:
        images  = images.to(device)
        outputs = model(images).squeeze(1)   # (batch, 1) → (batch,) para comparar con labels
        all_preds.extend(outputs.cpu().numpy())   # guardar predicciones
        all_labels.extend(dias.numpy())            # guardar valores reales

# Convertir listas a arrays de numpy para calcular métricas
all_preds  = np.array(all_preds)
all_labels = np.array(all_labels)

# Calcular e imprimir todas las métricas
imprimir_metricas(all_labels, all_preds, n_features=1)

# Gráfico de días predichos vs días reales
plt.figure(figsize=(7, 6))
plt.scatter(all_labels, all_preds, alpha=0.4, color="steelblue",
            edgecolors="k", linewidths=0.3)
plt.plot([all_labels.min(), all_labels.max()],
         [all_labels.min(), all_labels.max()],
         "r--", linewidth=2, label="Predicción perfecta")
plt.xlabel("Días Reales")
plt.ylabel("Días Predichos")
plt.title("EfficientNet-B3 — Días Reales vs Predichos")
plt.legend()
plt.tight_layout()
plt.savefig(f"{BASE}/Dataset CNN/Regresión/EfficientNet-B3/Regresion_predicho_vs_real_con_AD.png",
            dpi=150, bbox_inches="tight")
plt.show()
```
